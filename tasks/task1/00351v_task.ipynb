```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np
```

```python
filename ="00351v.jpg"

img = cv.imread(filename, cv.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
```

```python
h,w = img.shape
h = int(h/3)

b = img[0:h,:]
g = img[h:2*h,:]
r = img[2*h:3*h,:]

plt.subplot(1,3,1)
plt.imshow(b[150:250,100:200], cmap="gray")

plt.subplot(1,3,2)
plt.imshow(g[150:250,100:200], cmap="gray")

plt.subplot(1,3,3)
plt.imshow(r[150:250,100:200], cmap="gray")
```

```python
k=np.zeros((5,5))
k[0,2]=1
g_ = cv.filter2D(g, cv.CV_8U, k)
g_ = cv.filter2D(g, cv.CV_8U, k)
r_ = cv.filter2D(r, cv.CV_8U,k)
r_ = cv.filter2D(r_, cv.CV_8U,k)
r_ = cv.filter2D(r_, cv.CV_8U,k)
r_ = cv.filter2D(r_, cv.CV_8U,k)
r_ = cv.filter2D(r_, cv.CV_8U,k)
r_ = cv.filter2D(r_, cv.CV_8U,k)

k=np.zeros((5,5))
k[2,0]=1
r_ = cv.filter2D(r_, cv.CV_8U,k)

plt.subplot(1,3,1)
plt.imshow(b[150:250,100:200],cmap="gray")
plt.subplot(1,3,2)
plt.imshow(g_[150:250,100:200],cmap="gray")
plt.subplot(1,3,3)
plt.imshow(r_[150:250,100:200],cmap="gray")
```

```python
stacked = np.dstack((b,g_,r_))
plt.imshow(cv.cvtColor(stacked, cv.COLOR_BGR2RGB))
```

```python
stacked = np.dstack((b,g_,r_))
plt.imshow(cv.cvtColor(stacked[150:250,100:200], cv.COLOR_BGR2RGB))
```
